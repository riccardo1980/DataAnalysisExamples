```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
# setup

def createImage(ax, sets, h = 2):
    t = np.linspace(-np.pi/2,3/2*np.pi, 600)
    x = np.cos(t)
    y = np.sin(t)

    ax.axis('tight')
    ax.axis('equal')
    ax.set_xlim([-h,h])
    ax.set_ylim([-h,h])
    
    ax.axes.get_xaxis().set_visible(False)
    ax.axes.get_yaxis().set_visible(False)
    ax.axis('off')
    
    print_settings = {'TN': {'fillcolor': 'green',
                             'x': np.concatenate((x[x>0], [0, 0, h, h, 0])),
                             'y': np.concatenate((y[x>0], [-h, h, h, -h, -h])),
                             'xt': 0.1*h, 'yt': 4/5*h, 
                             'text': 'True\n Negatives', 'textcolor': 'black'},
                      'FN': {'fillcolor': 'blue',
                             'x': np.concatenate(([0, -h, -h, 0], x[x<0])),
                             'y': np.concatenate(([-h, -h, h, h], y[x<0])),
                             'xt': -0.9*h, 'yt': 4/5*h,
                             'text': 'False\n Negatives', 'textcolor': 'lightblue'},
                      'FP': {'fillcolor': 'lightgreen',
                             'x': x[x>0], 'y': y[x>0],
                             'xt': h/7, 'yt': 0, 
                             'text': 'False\n Positives', 'textcolor': 'black'},
                      'TP': {'fillcolor': 'lightblue',
                             'x': x[x<0], 'y': y[x<0],
                             'xt': -h/2, 'yt': 0, 
                             'text': 'True\n Positives', 'textcolor': 'black'}
            }
    
    for s in sets:
        ax.fill(print_settings[s]['x'],print_settings[s]['y'], 
                 print_settings[s]['fillcolor'])
        ax.text(print_settings[s]['xt'],print_settings[s]['yt'], 
                 print_settings[s]['text'], color= print_settings[s]['textcolor'])


```

```python
# Accuracy
f, (num, den) = plt.subplots(1,2)
createImage(num, ['TP', 'TN'])
createImage(den, ['TP', 'TN', 'FP', 'FN'])
```

```python
# Precision
f, (num, den) = plt.subplots(1,2)
createImage(num, ['TP'])
createImage(den, ['TP', 'FP'])
```

```python
# Recall
f, (num, den) = plt.subplots(1,2)
createImage(num, ['TP'])
createImage(den, ['TP', 'FN'])
```
